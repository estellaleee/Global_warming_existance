```python
weather["snow_depth"] = DF_snow["snow_depth"]
```

```python
weather["tmax"] = DF_hot["Tmax"]
```

# Groupby

```python
tmax = DF_hot.groupby(["year"])["Tmax"].mean()
```

```python
tmin = DF_cold.groupby(["year"])["Tmin"].mean()
```

```python
snow_depth = DF_snow.groupby(["year"])["snow_depth"].mean()
```

```python
weather_final = pd.concat([tmax, tmin, snow_depth], axis=1)
```

```python
weather_final.reset_index(inplace=True)
```

### TMAX = Maximum temperature (tenths of degrees C)
### TMIN = Minimum temperature (tenths of degrees C)

```python
weather_final['Tmax'] = weather_final['Tmax'] * 0.1
weather_final['Tmin'] = weather_final['Tmin'] * 0.1
```
